## Load in and display data

```python
def plot_biom_data_dis(biom_data):
    SampleX = biom_data.matrix_data.transpose().astype('int')
    # Details on Taxa data
    Taxa_max_reads = sp.sparse.coo_matrix.todense(sp.sparse.csr_matrix.max(SampleX,0))
    Taxa_mean_reads = sp.sparse.csr_matrix.mean(SampleX,0)
    Taxa_total_reads = sp.sparse.csr_matrix.sum(SampleX,0)

    #Details on Location data
    Loc_max_reads = sp.sparse.coo_matrix.todense(sp.sparse.csr_matrix.max(SampleX,1))
    Loc_mean_reads = sp.sparse.csr_matrix.mean(SampleX,1)
    Loc_total_reads = sp.sparse.csr_matrix.sum(SampleX,1)
    
    # Plot distribution of counts at 
    plot_microbes.log_hist_plot(Loc_total_reads, 
                                title='Histogram of species counts at each location', 
                                xlabel='Total counts per location', 
                                ylabel='Number of locations')
    plot_microbes.log_hist_plot(Taxa_total_reads.transpose(), 
                                title='Total counts for each species', 
                                xlabel='Total counts per species', 
                                ylabel='Number of species')
```
